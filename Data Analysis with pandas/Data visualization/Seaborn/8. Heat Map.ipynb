# Heatmap

- Heatmap is one of a very useful visualization techniques. You can use this technique to see correlations between numerical variables. Letâ€™s use the corr method to see this.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Creating an array

ary = np.arange(10,42,2).reshape(4,4)
```

```python
ary
```

```python
#Visualising the array in heatmap

sns.heatmap(ary, annot=True, linewidths=2, linecolor= "y")
```

```python
dfair=sns.load_dataset('flights')
```

```python
dfair
```

```python
#Creating pivot tabel of count of passenges based on year and month 

flight = dfair.pivot_table( index = "year", columns = "month", values = "passengers", aggfunc="sum")
```

```python
flight
```

```python
#Creating a heatmap

sns.heatmap(flight)
```

```python
#Creating a heatmap with same values in the pivot tabel created before

sns.heatmap(flight, linewidths=1, linecolor="y", annot=True, fmt="d")
plt.xticks(rotation = 45)
plt.show()
```

```python
#Using cmap parameter we can change the color of heat map

sns.heatmap(flight, linewidths=1, linecolor="y", annot=True, fmt="d", cmap="summer")
plt.xticks(rotation = 45)
plt.show()
```

```python
sns.heatmap(flight, linewidths=1, linecolor="y", annot=True, fmt="d", cmap="winter")
plt.xticks(rotation = 45)
plt.show()
```

## Corelation 

```python
dfair.corr()
```

```python
#Creating heatmap with corelation values

sns.heatmap(dfair.corr())
```

- You can see the relationship between the numerical variables in this graph. You can also use the annot parameter to see the numeric values in each cell.

```python
sns.heatmap(dfair.corr(),annot=True)
```
