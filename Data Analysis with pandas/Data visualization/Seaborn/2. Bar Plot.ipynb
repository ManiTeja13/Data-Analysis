# Bar plot

- A bar plot represents an estimate of the central tendency for a numeric variable with the height of each rectangle.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dftips = pd.read_csv("data\\tips.csv")
```

```python
dftips.columns
```

```python
dftips.columns = ['total', 'tip', 'gender', 'smoker', 'day', 'time', 'size']
```

```python
dftips
```

```python
#Creating a bar plot total bill day wise

sns.barplot(x = dftips["day"], y = dftips["total"])
```

```python
#Here axis decides the vertical and horizontal view of bar plot

sns.barplot(y = dftips["day"], x = dftips["total"])
```

```python
#Creating a bar plot total bill day wise and comparing it with time(using hue parameter)

sns.barplot(x = dftips["day"], y = dftips["total"], hue = dftips["time"])
```

```python
#Creating a bar plot total bill day wise and comparing it with genders(using hue parameter)

sns.barplot(x = dftips["day"], y = dftips["total"], hue = dftips["gender"])
```

```python
sns.barplot(x = "day", y = "total", hue = "gender", data = dftips)
```

```python
#Changing color 

sns.barplot(x = "day", y = "total", hue = "gender", data = dftips, color="b")
```

```python
#using palette

sns.barplot(x = "day", y = "total", hue = "gender", data = dftips, palette='cool')
```

```python
sns.barplot(x = "day", y = "total", hue = "gender", data = dftips, palette='Wistia')
```
