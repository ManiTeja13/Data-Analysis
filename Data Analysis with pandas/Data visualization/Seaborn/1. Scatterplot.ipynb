# Scatter plot

- Scatter plot shows the data in dots represent with x and y axis.

- The best technique for understanding data is the scatter plot. The scatter plot is used to display the relationship between two variables.

- hue parameter in `sns.scatter()` will produce points with different colors. Can be either categorical or numeric, although color mapping will behave differently in latter case.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dfship = pd.read_csv("data\\titanic.csv")
```

```python
dfship
```

```python
dfship.columns
```

```python
#Creating Scatter plot to find the relation between  age and fare

sns.scatterplot(x = "age", y = "fare", data = dfship)
```

```python

plt.figure(figsize=(16,8))                                 #increasing figure size
sns.scatterplot(x="age", y="fare", data = dfship)
```

```python
#Creating Scatter plot to find the relation between  age and fare and 'hue' as gender

sns.scatterplot(x="age", y="fare", data = dfship, hue="Gender")
```

**Insight**

- As in above plot we can observe that their and more Male compared to Female and 21 to 40 age people are more. 

- Most of the Male's fare is bellow 100 and 2 are above 500.

```python
#Creating Scatter plot to find the relation between  age and fare and 'hue' as gender and style as who 

sns.scatterplot(x ="age", y="fare", data = dfship, hue = "Gender", style = "who")
```

- In the above plot style(markers) is to differentiate the category of who(man,woman,child) 

```python
#Changing markers size 

sns.scatterplot(x ="age", y="fare", data = dfship, hue = "Gender", style = "who", size = "who", sizes = (100,300) )
```

```python
plt.figure(figsize=(16,8))
sns.scatterplot(x ="age", y="fare", data = dfship, hue = "Gender", style = "who", size = "who", sizes = (100,300) )
```

```python
#Creating Scatter plot to find the relation between  who and fare and 'hue' as alive 

sns.scatterplot(x ="who", y="fare", data = dfship, hue = "alive")
```

```python
#palette used to change the theme of the colors

sns.scatterplot(x ="who", y="fare", data = dfship, hue = "alive", palette = 'ocean')
```
