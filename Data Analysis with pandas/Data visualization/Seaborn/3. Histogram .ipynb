# Histogram

- Histogram shows the distribution of the data. You can use the histogram plot to see the distribution of one or more variables.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dftips = pd.read_csv("data\\tips.csv")
```

```python
dftips.columns = ['total', 'tip', 'gender', 'smoker', 'day', 'time', 'size']
```

```python
#Creating Histogram(dristribution) of Total bill

sns.histplot(dftips["total"])
```

```python
#hist using pandas

plt.hist(dftips["total"])
plt.show()
```

```python
#Creating Histogram(dristribution) of Total bill with legend

sns.histplot(dftips["total"], label = "Total Bill")
plt.legend()
```

```python
#Creating Histogram(dristribution) of size

sns.histplot(dftips["size"])
```

```python
#Creating Histogram(dristribution) of Tips with kde

sns.histplot(dftips["tip"],kde=True)
```

```python
#Creating Histogram(dristribution) of total bill with kde

sns.histplot(dftips["total"],kde=True)
```

```python
#Creating Histogram(dristribution) of total bill with hue

sns.histplot(data=dftips,x="total",hue="gender")
```

```python
#Creating Histogram(dristribution) of total bill with kde and hue

sns.histplot(data=dftips,x="total",hue="gender",kde=True)
```

```python
#using bins

sns.histplot(dftips["total"], bins = 50)
```

```python
#Changing color of hist

sns.histplot(dftips["total"], bins = 25, kde = False, color="r")
```
