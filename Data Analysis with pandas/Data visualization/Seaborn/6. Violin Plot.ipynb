# Violin Plot

- You can think of the violin plot as a box plot. This plot is used to compare the distribution of numerical values among categorical variables. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dfmall = pd.read_csv("data\\supermarket_sales.csv")
```

```python
dfmall
```

```python
dfmall["Total"]
```

```python
#Creating violin plot for total feature

sns.violinplot(y = "Total", data = dfmall)
```

```python
#Creating violin plot for total feature with width adjusted

sns.violinplot(y = "Total", data = dfmall, width=0.5)
```

```python
#Changing the color of violin

sns.violinplot(y = "Total", data = dfmall, width=0.5, color="yellow")
```

```python
#Horizontal representation

sns.violinplot(x = "Total", data = dfmall,width=0.5, color="r")
```

```python
sns.violinplot(x="Payment", y="Total", data=dfmall, hue="City")
```

```python
plt.figure(figsize=(16,8))
sns.violinplot(x="Payment", y="Total", data=dfmall, hue="Gender")
```

```python
#Here split parmeter shows the distribution of hue category separate

plt.figure(figsize=(16,8))
sns.violinplot(x="Payment", y="Total", data=dfmall, hue="Gender", split=True)
```

```python
#Using inner parameter we can also see the Quartiles of the data

plt.figure(figsize=(16,8))
sns.violinplot(x="Payment", y="Total", data=dfmall, hue="Gender", split=True, inner='quartile')
```
