# Box Plot

- A box-plot is a very useful and standardized way of displaying the distribution of data based on a five-number summary (minimum, first quartile, second quartile(median), third quartile, maximum). It helps in understanding these parameters of the distribution of data and is extremely helpful in detecting outliers.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dfmall = pd.read_csv("data\\supermarket_sales.csv")
```

```python
dfmall
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
#Detecting Outliers in Total feature

sns.boxplot(dfmall["Total"])
```

```python
#Horizontal representation

sns.boxplot(x = "Total", data= dfmall)
```

```python
#Width parameter is for the width of the box

sns.boxplot(x = "Total", data= dfmall, width=0.8)
```

```python
sns.boxplot(y = "Total", data= dfmall, width=0.3)
```

```python
#Using hue parameter

sns.boxplot(x = "Payment", y = "Total", data = dfmall, hue = "Gender")
```

```python
#showmeans is a parameter which shows mean value in the plot

plt.figure(figsize=(16,8))
sns.boxplot(x = "Payment", y = "Total", data = dfmall, hue = "Gender", showmeans = True)
```

```python
#Using palette in Box plot

sns.boxplot(x = "Total", y = "Payment", data = dfmall, hue = "Gender", showmeans = True, palette='Spectral' )
```

```python
#Box plot for all numerical features in the dataframe

plt.style.use("ggplot")
plt.figure(figsize=(18,8))
sns.boxplot(data = dfmall)
```
