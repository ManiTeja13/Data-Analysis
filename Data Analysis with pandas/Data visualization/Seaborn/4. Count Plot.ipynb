# Count plot

- Count plot is used to give the count of the category in the Categorical variable | Discrete Variable in the Bar. It is same as bar plot.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dfmall = pd.read_csv("data\\supermarket_sales.csv")
```

```python
dfmall
```

**All the categorical Variables in the above dataframe**

- Branch
- City
- Customer type
- Gender
- Product_line
- Payment

```python
#Creating Count plot for the count of City

sns.countplot(x = "City", data = dfmall )
```

```python
#Creating Count plot for the count of product line with figure size

plt.figure(figsize=(16,8))
sns.countplot(x = "Product line", data = dfmall )
```

```python
dfmall["Product line"].value_counts()
```

```python
#In Count plot axis decides the (vertical and horizontal) resprentation

plt.figure(figsize=(16,8))
sns.countplot(y = "Product line", data = dfmall )
```

```python
#We Can use hue parameter in count plot

plt.figure(figsize=(16,8))
sns.countplot(x = "Product line", data = dfmall, hue="Gender")
```

```python
#Giving hue order with hue_order parameter

plt.figure(figsize=(16,8))
sns.countplot(x = "Product line", data = dfmall, hue="Gender",hue_order=["Male", "Female"])
```

```python
#Using palette

plt.figure(figsize=(16,8))
sns.countplot(x = "Product line", data = dfmall, hue="Gender", palette = 'rocket' )
```
