# Pair plot

- Seeing the pair relationship between the variables in the dataset is one of the important steps of data analysis. You can use the pairplot method to see the pair relations of the variables. This function creates cross-plots of each numeric variable in the dataset. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_iris=sns.load_dataset("iris")
```

```python
df_iris
```

```python
#Creating pair plot 

sns.pairplot(df_iris)
```

```python
#Using hue parameter

sns.pairplot(df_iris,hue="species")
```

```python
#Changing the height of the plots

sns.pairplot(df_iris, hue="species", height=3)
```

- Since the variables are numerical, a probabilistic density function is automatically drawn on the diagonal axis of the graph. You can use the diag_kind parameter to draw histograms on the diagonal axis.

```python
sns.pairplot(df_iris, hue="species", diag_kind="hist")
```
