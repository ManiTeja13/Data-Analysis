## Plotting with Pandas

- For a quick Visualization we no need to use visualization libraries like matplotlib and seaborn. `pandas will provide some quick visualization plots`. 

- Syntax - `Dataframe['column'].plot(kind = "")`
  - `Datraframe['column'].plot.plotname`
  
  
- Common parameters in `.plot` are x,y,figsize,title,xticks,yticks,legend etc.... some are based on plots.


- The kind of plot to produce:

1. `line` : line plot (default) 
2. `bar`: vertical bar plot
3. `barh` : horizontal bar plot
4. `hist` : histogram
5. `box` : boxplot
6. `kde` : Kernel Density Estimation plot
7. `density` : same as ‘kde’
8. `area` : area plot
9. `pie` : pie plot
10. `scatter` : scatter plot
11. `hexbin` : hexbin plot

```python
import pandas as pd
```

```python
df=pd.read_csv("data\Iris.csv")
```

```python
df
```

**Line plot**

- used to display a frequency of data along a line segment

```python
df["SepalLengthCm"].plot(kind="line",figsize=(10,5),c="r",ylabel="SepalLengthCm")
```

```python
df["SepalLengthCm"].plot.line()
```

**Bar plot**

- used to display categorical data

- Use the .bar() method with optional argument stacked to display data in columns that are positioned on top of one another.

```python
df["Species"].value_counts().plot(kind="bar",color='r')
```

```python
df["Species"].value_counts().plot(kind="barh",color='r')         #horizontal representation of bar
```

**Histogram plot**

- used to display the distribution of a numerical dataset

- Use the `.hist()` method with the bins argument to generate a histogram plot. The bins specify the amount of buckets we want to allocate to split the data into for review.

```python
df["SepalLengthCm"].plot(kind="hist",bins=50,edgecolor="k")
```

**Box plot**

- used to display distribution, central value and variability of data and `outliers`

- Use the .box() method to generate basic box plots

```python
df["SepalLengthCm"].plot(kind="box")
```

```python
df.iloc[:,1:].plot(kind="box")
```

**KDE (Kernel Density Estimation) plot**

- used for data smoothing and to display the density distribution of the data

```python
df["SepalLengthCm"].plot(kind="kde")
```

**Area plot**

- used to display quantitative

- Use the `.area()` method to generate an area plot. The area plot may be thought of as a line plot with the area underneath filled in.

```python
df.loc[:,["SepalLengthCm","SepalWidthCm"]].plot(kind="area")
```

**Scatter plot**

- used to see relationship of two columns

```python
df.plot(x="SepalLengthCm",y="SepalWidthCm",kind="scatter")
```

**Pie chart**

- used to see the percentage of each value in a categorical column

```python
df["Species"].value_counts().plot(kind="pie",autopct='%0.2f%%')
```

**Hexbin plot**

- essentially scatter plots displayed as hexagons based on data point density

- Use the `.hexbin()` method to gain insight on the density of data from a alternative perspective. The data at the blue end of the spectrum is less dense than that of the red. Or, there were more data points located around the red hexagon than that of the blue areas.

```python
df.plot(kind="hexbin",x="SepalLengthCm",y="SepalWidthCm",cmap="coolwarm")
```
