**Exploring Parameters in `read_csv` function**

```python
pd.read_csv('data/std_data1.csv')       #reading std_data1.csv
```

**1. usecols**

```python
#Parameter -> usecols --- will only take specified
```

```python
pd.read_csv('data/std_data1.csv',usecols=["ENGLISH"])   #using column name  
```

```python
pd.read_csv('data/std_data1.csv',usecols=[2])   #using index of the column
```

```python
pd.read_csv('data/std_data1.csv',usecols=["ENGLISH","SCIENCE"])   #multiple columns
```

**2. nrows**

```python
pd.read_csv('data/std_data1.csv',nrows=4)     #returns specified rows
```

**3. skiprows**

```python
pd.read_csv('data/std_data2.csv') 
```

```python
pd.read_csv('data/std_data2.csv',skiprows=1) #it will skip specified rows from starting index 
```

```python
pd.read_csv('data/std_data2.csv',skiprows=2)
```

```python
pd.read_csv('data/std_data2.csv',skiprows=[2])  
#when the spedcified value in [list] it will skips only that row
```

```python
pd.read_csv('data/std_data2.csv',skiprows=[1,7,3])
```

**4. header**

```python
pd.read_csv('data/std_data_nocols.csv')
```

```python
pd.read_csv('data/std_data_nocols.csv',header=None)   
#column names will be index values when hearder=None
```

**5. names**

```python
pd.read_csv('data/std_data_nocols.csv',header=None,names=["Name","Cls","Eng","Sci","Comp"]) 
#passing column names
```

**6. index_col**

```python
pd.read_csv('data/std_data1.csv')   
```

```python
pd.read_csv('data/std_data1.csv',index_col="SNAME")   
#changing index by specified column
```
